## Data Preprocessing

```python
print('Before removing duplicate values: ')
print('No. of Rows: ',df.shape[0])
print('No. of Columns: ',df.shape[1])

df.drop_duplicates(inplace = True)

print('\nAfter removing duplicate values: ')
print('No. of Rows: ',df.shape[0])
print('No. of Columns: ',df.shape[1])
```

```python
df['rate'].unique()
```

df['rate'] >>

    i) Here, dtype for 'rate' is object. I have to convert it into the 
    integer value.

    ii) I have to remove '/5' from each of the samples so that I could get
    only initial float value.

    iii) It also contain some '-','NEW' values. So, I have to remove those.
    
    iv) Also, I have to remove the NaN values from the 'rate' attributes.

```python
def rate_process(i):
    if (i == 'NEW') or (i == '-') or (i == 'nan'):
        return np.nan 
    else:
        i = str(i).split('/')
        return float(i[0])
```

```python
df['rate'] = df['rate'].apply(rate_process)
df['rate'].unique()
```

```python
df['rate'].isnull().sum()
```

Type 1: To remove commas from the integer value using str.replace() function

```python
def comma_process(i):
    i = str(i)
    
    if ',' in i:
        i = i.replace(',', '')
        return float(i)
    else:
        return float(i)
```

def comma_process will get a value from .apply fucntion. If the ',' comma is present in the value them replace that comma with noting and return its float value to the function. On the other hand, if commma is not present then return its float value only.

```python
df['cost2ppl'] = df['cost2ppl'].apply(comma_process)
df['cost2ppl'].unique()
```

Type 3: To remove commas from the integer value using anonymous function.

```python
method_3 = df['cost2ppl'].astype(str)
method_3
```

So, this is how I could remove the commas from any integer values by taking anonymous function.
